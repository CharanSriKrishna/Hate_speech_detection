```python
import numpy as np 
import pandas as pd
import os
import chardet
```

```python
dataset=pd.read_csv('annotations_metadata.csv')
```

```python
dataset.head()
```

```python
ans={'ISO-8859-9':0, 'Windows-1252':0, 'Windows-1254':0, 'MacRoman':0, 'ascii':0, 'EUC-JP':0, 'utf-8':0, 'ISO-8859-1':0}
file_path='data\\all_files'
for i in range(0,len(dataset)):
    addreass = dataset['file_id'][i]+'.txt'
    file=os.path.join(file_path, addreass)
    with open(file, 'rb') as file:
        rawdata = file.read(100)
    result = chardet.detect(rawdata)
    print(result['encoding'])
    ans[result['encoding']]+=1
    #with open(file, 'r', encoding=result['encoding']) as file:
    #    contents = file.read()
    #file.close()
    #print(contents)
    print(i)
print(ans)
```

```python
file_path='data\\all_files'
for i in range(0,len(dataset)):
    addreass = dataset['file_id'][i]+'.txt'
    file=os.path.join(file_path, addreass)
    with open(file, 'rb') as file:
        rawdata = file.read(100)
    result = chardet.detect(rawdata)
    if(result['encoding'] != 'ascii'):
        continue
    file.close()
    with open(file, 'r') as file:
        contents = file.read()
    file.close()
    print(contents)
    print(i)
print(ans)
```

```python
file_path = 'data\\all_files'

for i in range(0,len(dataset)):
    address = dataset['file_id'][i] + '.txt'
    file = os.path.join(file_path, address)
    
    with open(file, 'rb') as file:
        rawdata = file.read(100)
    
    result = chardet.detect(rawdata)
    
    if(result['encoding'] != 'ascii'):
        continue
    
    file.close()
    with open(file, 'r' ,encoding='ascii') as file:
        contents = file.read()
    
    file.close()
    print(contents)
    print(i)
    
```
